Im Folgenden berechnen wie mittels der generalized pareto distribution (GPD) die Kennzahlen f체r das Experiment im Paper.

## Wir ermitteln p_0

```python
import numpy as np
from scipy.stats import genpareto

# Parameter der Generalized Pareto Distribution
shape = 0.5  # Formparameter (k)
scale = 775  # Skalierung (sigma)
loc = 350    # Verschiebung (theta)

# Anzahl der Stichproben
n_samples = 1000000

# Stichproben generieren
samples = genpareto.rvs(c=shape, scale=scale, loc=loc, size=n_samples)

# Werte auf die Range [350, 2500] beschr채nken
samples = np.clip(samples, 350, 2500)

# T_0 ist der Mittelwert der Stichprobe
T0 = np.mean(samples)
p0 = 1 / T0

print("T_0 ist:", round(T0))
print("p_0 ist:", round(p0, 4))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn-Stil setzen
sns.set_theme(style="whitegrid")

# Histogramm der Stichproben plotten
sns.histplot(samples, bins=50, kde=True, color='g', stat="density", alpha=0.6)

# Titel und Achsenbeschriftungen hinzuf체gen
plt.title('Histogramm der Generalized Pareto Distribution', fontsize=14)
plt.xlabel('Jahre', fontsize=12)
plt.ylabel('Dichte', fontsize=12)

plt.show()
```

## Wir ermitteln p_1

```python
import numpy as np
from scipy.stats import genpareto

# Parameter der Generalized Pareto Distribution
shape = 0.5  # Formparameter (k)
scale = 13  # Skalierung (sigma)
loc = 100    # Verschiebung (theta)

# Anzahl der Stichproben
n_samples = 1000000

# Stichproben generieren
samples = genpareto.rvs(c=shape, scale=scale, loc=loc, size=n_samples)

# Werte auf die Range [100, 1000] beschr채nken
samples = np.clip(samples, 100, 1000)

# p_0 ist der Mittelwert der Stichprobe
T1 = np.mean(samples)
p1 = 1 / T1

print("T_1 ist:", round(T1))
print("p_1 ist:", round(p1, 4))
```

```python
# Histogramm der Stichproben plotten
sns.histplot(samples, bins=50, kde=True, color='g', stat="density", alpha=0.6)

# Titel und Achsenbeschriftungen hinzuf체gen
plt.title('Histogramm der Generalized Pareto Distribution', fontsize=14)
plt.xlabel('Jahre', fontsize=12)
plt.ylabel('Dichte', fontsize=12)

plt.show()
```

### Wir berechnen PS und PN

```python
ps = 1-(float(p0)/float(p1))
print("PS ist:", round(ps, 4))

pn = 1-((1-p1)/(1-p0))
print("PN ist:", round(pn, 4))
```
